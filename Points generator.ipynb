```python
import math
import random
import numpy as np
import pandas as pd
from math import pi, log
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from mpl_toolkits.mplot3d import Axes3D
```

# Generate data using sklearn

```python
features, target = make_regression(n_samples = 5000,
                                   n_features = 2,
                                   n_informative = 2,
                                   n_targets = 1,
                                   noise = 0.0,
                                   random_state = 0)
```

```python
data = np.hstack((features,target.reshape(-1,1)))
```

```python
x = data[:,0]
y = data[:,1]
z = data[:,2]
sum(x*x + y*y + z*z <= 50*50)
```

```python
lables = x*x + y*y + z*z <= 50*50
data = np.hstack((data,lables.reshape(-1,1)))
```

```python
%matplotlib inline
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data[data[:,3]<=0][:,0], data[data[:,3]<=0][:,1],data[data[:,3]<=0][:,2])
plt.show()
```

```python
df = pd.DataFrame(data, columns=["x", "y", "z", "in_sphere"])
```

```python
df_train = df[:-1000]
df_train.to_csv("data_train.csv", index=False)
```

```python
df_test = df[-1000:]
df_test.to_csv("data_test.csv", index=False)
```

# TEST

```python
points = []
```

```python
r = 7
number_of_points = 1000
for x in range(number_of_points):
    x = round(random.uniform(-10, 10), 2)
    y = round(random.uniform(-10, 10), 2)
    d = math.sqrt(math.pow((x-0), 2) + math.pow((y-0), 2))
    if d > r:
        points.append([x, y, False])
    else:
        points.append([x, y, True])
```

```python
points = np.array(points)
```

```python
plt.scatter(points[:,0], points[:,1])
plt.show()
```

```python
df = pd.DataFrame(points, columns=["x","y","in_circle"])
```

```python
df.to_csv("points.csv", index=False)
```
